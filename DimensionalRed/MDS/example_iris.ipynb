```python
from sklearn.datasets import load_iris
data= load_iris()
X,y = data['data'],data['target']
```

```python
X,y
```

```python
colors = []
for i in range(len(y)):
    if y[i] == 1:
        colors.append('#FFB300')
    elif y[i] == 2:
        colors.append('#E64A19')
    else:
        colors.append('#00FF00')
```

```python
mds = MDS(X,2)
```

```python
mds.project()
```

```python
mds.plot_scores_2d(colors = colors, dim_1 = 1, dim_2 = 2)
```

```python
from sklearn.manifold import MDS
embedding = MDS(n_components=2)
X_transformed = embedding.fit_transform(X)
X_transformed
```

```python
from sklearn.manifold import MDS
embedding = MDS(n_components=2)
X_transformed = embedding.fit_transform(X)
X_transformed
```

```python
def plot_scores_2d(data, colors, grid = True, dim_1 = 1, dim_2 = 2):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(15,10))
    plt.axhline(c = 'black', alpha = 0.2)
    plt.axvline(c = 'black', alpha = 0.2)
    plt.scatter(data.T[dim_1 - 1,:], data.T[dim_2 - 1,:], c = colors)
    plt.grid(grid)
    plt.title('Multi-Dimensional Scaing')
    plt.xlabel('${}^a$ dimension'.format(dim_1))
    plt.ylabel('${}^a$ dimension'.format(dim_2))
    plt.show()
```

```python
plot_scores_2d(X_transformed,colors = colors,grid = True, dim_1 = 1, dim_2 = 2)
```
