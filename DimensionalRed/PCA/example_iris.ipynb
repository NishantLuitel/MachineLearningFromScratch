```python
from sklearn.datasets import load_iris
from pca import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data= load_iris()
X,y = data['data'],data['target']
X,y
```

```python
p = PCA(X,num_dim = 2)
p.project()
```

```python
colors = []
for i in range(len(y)):
    if y[i] == 1:
        colors.append('#FFB300')
    elif y[i] == 2:
        colors.append('#E64A19')
    else:
        colors.append('#00FF00')
```

```python
p.plot_scores_2d(colors = colors)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
p_c= pca.fit_transform(X)

```

```python
def plot_scores_2d(data, colors, grid = True, dim_1 = 1, dim_2 = 2):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(15,10))
    plt.axhline(c = 'black', alpha = 0.2)
    plt.axvline(c = 'black', alpha = 0.2)
    plt.scatter(data.T[dim_1 - 1,:], data.T[dim_2 - 1,:], c = colors)
    plt.grid(grid)
    plt.title('Principal Component Analysis')
    plt.xlabel('{}st principal component'.format(dim_1))
    plt.ylabel('{}nd principal component'.format(dim_2))
    plt.show()
```

```python
plot_scores_2d(p_c,colors = colors,grid = True, dim_1 = 1, dim_2 = 2)
```
