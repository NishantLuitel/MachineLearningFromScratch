# LinearRegression Using both Scikit and Custom model(Without Bias)

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X,y = data['data'],data['target']


print(X.shape,y.shape)
Xt,yt = X[:350],y[:350]
Xe,ye = X[350:],y[350:]
```

```python
y
```

```python
lr = LR(fit_intercept=False)
lr.fit(Xt,yt)
```

```python
yp = lr.predict(Xe)
```

```python
ye , yp
```

```python
lr_ = LinearRegression(Xt,yt,use_bias=False)
lr_.fit()
```

```python
yt.shape,Xt.shape
```

```python
yp_ = lr_.predict(Xe)
```

```python
ye,yp_ 
```

```python
lr_.weights()
```

# LinearRegression Using both Scikit and Custom model(With Bias)

```python
#Scikit Tool

lr = LR()
lr.fit(Xt,yt)

yp = lr.predict(Xe)
```

```python
ye , yp
```

```python
#Custom model

lr_ = LinearRegression(Xt,yt)
lr_.fit()

yp_ = lr_.predict(Xe)
```

```python
ye,yp_ 
```
