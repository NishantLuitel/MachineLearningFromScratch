```python
import numpy as np
from linear_models import Lasso
```

```python
from sklearn.linear_model import Lasso as L
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X,y = data['data'],data['target']


print(X.shape,y.shape)
Xt,yt = X[:350],y[:350]
Xe,ye = X[350:],y[350:]
```

```python
alpha = 0.5
y
```

```python
lr = L(alpha = alpha)
lr.fit(Xt,yt)
```

```python
yp = lr.predict(Xe)
```

```python
ye , yp
```

```python
import matplotlib.pyplot as plt

def plot_line_y_only(y):
    x = np.arange(len(y))
    plt.plot(x,y)
    plt.show()
    
    
print("Predicted:",end = "\n-----------\n")    
plot_line_y_only(yp)

print("True:",end = "\n--------\n")
plot_line_y_only(ye)
```

```python
#Now lets use our custom model


lr_ = Lasso(Xt,yt,alpha = alpha,no_iter = 3000,learning_rate = 0.1)
lr_.fit()
```

```python
yp_ = lr_.predict(Xe)
```

```python
ye,yp_ 
```

```python
print("Predicted:",end = "\n-----------\n")    
plot_line_y_only(yp_)

print("True:",end = "\n--------\n")
plot_line_y_only(ye)
```
