# Ridge regression Using both Scikit and Custom model

```python
from sklearn.linear_model import Ridge as R
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X,y = data['data'],data['target']


print(X.shape,y.shape)
Xt,yt = X[:350],y[:350]
Xe,ye = X[350:],y[350:]
```

```python
alpha = 0.5
y
```

```python
lr = R(alpha = alpha)
lr.fit(Xt,yt)
```

```python
yp = lr.predict(Xe)
```

```python
ye , yp
```

```python
import matplotlib.pyplot as plt

def plot_line_y_only(y):
    x = np.arange(len(y))
    plt.plot(x,y)
    plt.show()
    
    
print("Predicted:",end = "\n-----------\n")    
plot_line_y_only(yp)

print("True:",end = "\n--------\n")
plot_line_y_only(ye)
```

```python
#Now lets use our custom model


lr_ = Ridge(Xt,yt,alpha = alpha)
lr_.fit()
```

```python
yp_ = lr_.predict(Xe)
```

```python
ye,yp_ 
```

```python
print("Predicted:",end = "\n-----------\n")    
plot_line_y_only(yp_)

print("True:",end = "\n--------\n")
plot_line_y_only(ye)
```
