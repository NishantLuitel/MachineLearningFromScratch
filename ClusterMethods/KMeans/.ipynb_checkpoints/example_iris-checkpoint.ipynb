```python
from kmeans import Kmeans
```

```python


from sklearn.datasets import load_iris
data = load_iris()
X,y = data['data'],data['target']

# from sklearn import random_projection
# transformer = random_projection.GaussianRandomProjection()
# X_new = transformer.fit_transform(faces_3)

```

```python

kmeans = Kmeans(X,3,500)
p_labels= kmeans.cluster()
```

```python

colors_predicted = []
for i in range(len(p_labels)):
    if p_labels[i] == 1:
        colors_predicted.append('#FFB300')
    elif p_labels[i] == 2:
        colors_predicted.append('#E64A19')
    else:
        colors_predicted.append('#00FF00')
        
colors_labels = []
for i in range(len(y)):
    if y[i] == 1:
        colors_labels.append('#FFB300')
    elif y[i] == 2:
        colors_labels.append('#E64A19')
    else:
        colors_labels.append('#00FF00')
```

```python
#Plotting only first two dimensions

kmeans.plot_scores_2d(colors = colors_predicted)
kmeans.plot_scores_2d(colors = colors_labels)

#Both graphs fairly preserve the clusters though, colors don't agree
# Another thing to note here is out of 4 feature a iris data set , I have only used first two to plot,
# that is a cluster does have boundry outside of its immediate neighbourhood
```
