```python
from sklearn.datasets import fetch_olivetti_faces
import sys


faces = fetch_olivetti_faces('data/')
sys.path.insert(0, '../../DimensionalRed/Random_Projection')

from randomP import RandomP
from kmeans import Kmeans
```

```python
X,y = faces['data'],faces['target']
faces_3,labels_3 = X[:30],y[:30]
```

```python
rProj = RandomP(faces_3)
projected = rProj.project()
```

```python
k_means = Kmeans(projected,3,1000)
p_labels = k_means.cluster()
```

```python
print(p_labels),labels_3
```

```python

colors_predicted = []
for i in range(len(p_labels)):
    if p_labels[i] == 1:
        colors_predicted.append('#FFB300')
    elif p_labels[i] == 2:
        colors_predicted.append('#E64A19')
    else:
        colors_predicted.append('#00FF00')
        
colors_labels = []
for i in range(len(labels_3)):
    if labels_3[i] == 1:
        colors_labels.append('#FFB300')
    elif labels_3[i] == 2:
        colors_labels.append('#E64A19')
    else:
        colors_labels.append('#00FF00')
```

```python
print("Predicted Clusters")
k_means.plot_scores_2d(colors = colors_predicted)


print("True Clusters")
k_means.plot_scores_2d(colors = colors_labels)
```
